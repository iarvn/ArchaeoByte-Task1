```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(42)
num_samples = 1000

data = {
    'Soil_Moisture': np.random.uniform(20, 60, num_samples),
    'Temperature': np.random.uniform(25, 35, num_samples),
    'Nutrient_Level': 5 + 1.5 * np.random.randn(num_samples)
}

df = pd.DataFrame(data)
```

```python
# Split the dataset into features (X) and target variable (y)
X = df[['Soil_Moisture', 'Temperature']]
y = df['Nutrient_Level']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Create a linear regression model
model = LinearRegression()
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred)**0.5)
```

```python
# Visualize the results 
plt.scatter(X_test['Soil_Moisture'], y_test, color='black', label='Actual')
plt.scatter(X_test['Soil_Moisture'], y_pred, color='blue', label='Predicted')
plt.xlabel('Soil Moisture')
plt.ylabel('Nutrient Level')
plt.legend()
plt.show()
```

```python
# Visualize the results with a bar graph
plt.bar(range(len(y_test)), y_test, color='black', label='Actual')
plt.bar(range(len(y_test)), y_pred, color='blue', alpha=0.5, label='Predicted')
plt.xlabel('Sample')
plt.ylabel('Nutrient Level')
plt.legend()
plt.show()
```
